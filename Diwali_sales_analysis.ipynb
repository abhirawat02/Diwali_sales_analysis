```python
ax=sns.countplot(df,x="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_gen=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sales_gen
```

```python
sns.barplot(sales_gen,x="Gender",y="Amount")
plt.show()
```

```python
ax=sns.countplot(df,x="Age_Group", hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_age=df.groupby(["Age_Group"],as_index=False)["Amount"].sum().sort_values(by="Amount")
sales_age
```

```python
sns.barplot(sales_age,x="Age_Group",y="Amount")
plt.show()
```

```python
sns.kdeplot(df,x="Amount",hue="Zone",multiple="fill")
plt.show()
```

```python
sales_state=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(25, 10)})
sns.barplot(sales_state,x="State",y="Amount",palette="spring")
plt.show()
```

```python
orders_state=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False)
sns.barplot(orders_state,x="State",y="Orders",palette="hot")
sns.set(rc={"figure.figsize":(15, 5)})
plt.show()
```

```python
ax=sns.countplot(df,x="Marital_Status")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sns.countplot(df,x="Occupation",hue="Gender",palette="hot")
sns.set(rc={"figure.figsize":(21,15)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sns.countplot(df,x="Product_Category")
sns.set(rc={"figure.figsize":(25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
orders_state=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(orders_state,x="Product_Category",y="Amount",palette="hot")
sns.set(rc={"figure.figsize":(40,10)})
plt.show()
```
